```python
#showing the classes of dataset

import seaborn as sns
import matplotlib.pyplot as plt
sns.countplot(data=data, x='Class')
plt.show()
```

```python
#showing trasaction amount graph

plt.figure(figsize=(10, 5))
plt.subplot(1,2,1)
sns.histplot(data['Amount'], bins=40, kde=True)
plt.title('Trasaction Amount')
plt.xlabel('Amount')
```

```python
# checking the transaction time graph

plt.figure(figsize=(10, 5))
plt.subplot(1,2,2)
sns.histplot(data['Time'], bins=50, kde=True)
plt.title('Trasaction Time')
plt.xlabel('Time')
```
